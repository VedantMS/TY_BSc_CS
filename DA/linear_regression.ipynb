```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

Slip 1

```python
#   Create ‘Position_Salaries’ Data set. Build a linear regression  model by identifying  independent and
#   target  variable. Split the variables  into training and testing sets. then  divide the training and testing sets
#   into a 7:3 ratio, respectively and print them. Build a simple linear regression model.

position_salaries = {   'level': range(1, 11),
                        'salary': [45000, 50000, 60000, 80000, 110000, 150000, 200000, 300000, 500000, 1000000]
                    }

df = pd.DataFrame(position_salaries)
```

```python
x = np.array(df[['level']])     # Independent Variable
y = np.array(df[['salary']])    # Dependent Variable
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#   Create ‘Salary’ Data set . Build a linear regression model by identifying independent and target
#   variable. Split the variables into training and testing sets and print them. Build a simple linear regression
#   model for predicting purchases.

salary =    {   'salary': [45000, 50000, 60000, 80000, 110000, 150000, 200000, 300000, 500000, 1000000],
                'purchase_amount': [4500, 5000, 6000, 8000, 11000, 15000, 20000, 30000, 50000, 100000]
            }

df = pd.DataFrame(salary)
```

```python
x = np.array(df[['salary']])            # Independent Variable
y = np.array(df[['purchase_amount']])   # Dependent Variable
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
sal = float(input("Enter Salary: "))
```

```python
prediction = regressor.predict([[sal]])
print("Salary: ", sal, "\nPredicted Purchase Amount: ", prediction[0][0])
```

```python
#   Build a simple linear regression model for Fish Species Weight Prediction.

df = pd.read_csv('Fish.csv')

x = np.array(df[['Length1', 'Length2', 'Length3', 'Height', 'Width']])      #   Independent Variables
y = np.array(df[['Weight']])                                                #   Dependent Variable
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
len1 = float(input("Enter Length 1: "))
len2 = float(input("Enter Length 2: "))
len3 = float(input("Enter Length 3: "))
height = float(input("Enter Height: "))
weight = float(input("Enter Width: "))
```

```python
weight = regressor.predict([[len1, len2, len3, height, weight]])
print("Weight of Fish based on given parameters: ", weight[0][0])
```
