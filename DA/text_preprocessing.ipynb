```python
#   Consider any text paragraph. Preprocess the text to remove any special characters and digits.
#   Generate the summary using extractive summarization process

text = input("Enter Text: ")
print(text)
text = re.sub(r'[^a-z|A-Z|,|.]\s', '', text)
text
```

```python
stopWords = set(stopwords.words("english"))
```

```python
words = word_tokenize(text)
sentences = sent_tokenize(text)

word_freq = {}

for word in words:
    if word in stopWords:
        continue
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

word_freq
```

```python
sent_value = {}

for sent in sentences:
    for word, freq in word_freq.items():
        if word in sent:
            if sent in sent_value:
                sent_value[sent] += freq
            else:
                sent_value[sent] = freq
sent_value
```

```python
summary = ''

summary_sent = heapq.nlargest(4, sent_value, key = sent_value.get)
summary = ''.join(summary_sent)

summary
```

```python
#   Consider text paragraph. "So, keep working. Keep striving. Never give up. Fall down seven times, get
#   up eight. Ease is a greater threat to progress than hardship. Ease is a greater threat to progress than
#   hardship. So, keep moving, keep growing, keep learning. See you at work." Preprocess the text to remove
#   any special characters and digits. Generate the summary using extractive summarization process.


text = "So, keep working. Keep striving. Never give up. Fall down seven times, get up eight. Ease is a greater threat to progress than hardship. Ease is a greater threat to progress than hardship. So, keep moving, keep growing, keep learning. See you at work."

text = re.sub(r'[^a-z|A-Z|,|.]\s', '', text)
text
```

```python
words = word_tokenize(text)
sentences = sent_tokenize(text)

word_freq = {}

for word in words:
    if word in stopWords:
        continue
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

sent_value = {}

for sent in sentences:
    for word, freq in word_freq.items():
        if word in sent:
            if sent in sent_value:
                sent_value[sent] += freq
            else:
                sent_value[sent] = freq

summary = ''

summary_sent = heapq.nlargest(4, sent_value, key = sent_value.get)
summary = ''.join(summary_sent)

summary
```
